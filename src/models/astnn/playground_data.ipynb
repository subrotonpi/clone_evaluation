```python
def jsonl_to_df(jf):
    import json
    import pandas as pd
    with open(jf, 'r') as json_file:
        json_list = list(json_file)
    x = []
    for json_str in json_list:
        result = json.loads(json_str)
        x.append([result['idx'], result['func']])
    return pd.DataFrame(x, columns=['idx', 'func'])
```

```python
import javalang
import pandas as pd
def parse_program(func):
    tokens = javalang.tokenizer.tokenize(func)
    parser = javalang.parser.Parser(tokens)
    tree = parser.parse_member_declaration()
    return tree
```

```python
m = "/home/egk204/projects/clone-type-iv/data/gptcb/data.jsonl"
codes = jsonl_to_df(m)
```

```python
for i, v in codes.iterrows():
    # print(v.func)
    try:
        parse_program(v.func)
    except:
        print(v.idx)
        print(v.func)
```

```python
for c in codes.func:
    try:
        parse_program(c)
    except:
        print(c)
        break
```
