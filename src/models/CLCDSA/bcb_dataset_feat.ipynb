```python
import pandas as pd
feat_to_get =['path','variable', 'arguments', 'operators', 'expressions', 'loops','operands', 'exceptions', 'exception_clause', 'mccable_complex']
    
```

```python
mapping = pd.read_csv("/student/egk204/projects/clone-type-iv/data/bcb/mapping.csv", names=['fid', 'cid'])
mapping = {j.split('/')[-1].split('.')[0]:i for i, j in enumerate(mapping.fid)}
mapping
```

```python
bcb = pd.read_csv("/student/egk204/projects/clone-type-iv/src/cross/clcdsa_model/features/bcb/bcb/None_java_features.csv", names=feat_to_get)
```

```python
bcb['idx'] = [ mapping[p.split('/')[-1].split('.')[0]] for p in bcb.path]
```

```python
n = ['c1', 'c2', 'label']
train = pd.read_csv("/student/egk204/projects/clone-type-iv/data/bcb/train.csv", names=n)
test = pd.read_csv("/student/egk204/projects/clone-type-iv/data/bcb/test.csv", names=n)
valid = pd.read_csv("/student/egk204/projects/clone-type-iv/data/bcb/valid.csv", names=n)
```

```python
def get_feat_data(pairs, bcb):
    dset = []
    for ind, row in pairs.iterrows():
        # print(type(row[0]), row[1], row[2])
        try:
            x1 = bcb[bcb.idx==(row[0])][feat_to_get[1:]].values[0]
            x2 = bcb[bcb.idx==(row[1])][feat_to_get[1:]].values[0]
            label = row[2]
            t = list(x1)+ list(x2)+[label]
            # print(t)
        except:
            # x1 = [0]*9
            # x2 = [0]*9
            # label = 1
            # t = list(x1)+ list(x2)+[label]
            continue
        dset.append(t)
    return pd.DataFrame(dset)
```

```python
train = get_feat_data(train, bcb)
```

```python
test = get_feat_data(test, bcb)
```

```python
valid = get_feat_data(valid, bcb)
```

```python
train.to_csv("train_clcdsa.csv", header=None, index=False)
test.to_csv("test_clcdsa.csv", header=None, index=False)
valid.to_csv("valid_clcdsa.csv", header=None, index=False)
```
